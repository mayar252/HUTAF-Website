```python
import os
import shutil

# Define source and destination folders
source_folder = "/kaggle/working/Foul-4/train/images"
destination_folder = "/kaggle/working/Football-Foul-Detection-3/train/images"

# Ensure the destination folder exists
os.makedirs(destination_folder, exist_ok=True)

# Move images
for filename in os.listdir(source_folder):
    if filename.lower().endswith((".jpg", ".png", ".jpeg")):
        src_path = os.path.join(source_folder, filename)
        dest_path = os.path.join(destination_folder, filename)
        shutil.move(src_path, dest_path)
        print(f"Moved: {filename}")

print("All images moved successfully!")

import os
import shutil

# Define source and destination folders
source_folder = "/kaggle/working/Foul-4/train/labels"
destination_folder = "/kaggle/working/Football-Foul-Detection-3/train/labels"

# Ensure the destination folder exists
os.makedirs(destination_folder, exist_ok=True)

# Move text files
for filename in os.listdir(source_folder):
    if filename.lower().endswith(".txt"):
        src_path = os.path.join(source_folder, filename)
        dest_path = os.path.join(destination_folder, filename)
        shutil.move(src_path, dest_path)
        print(f"Moved: {filename}")

print("All text files moved successfully!")
```

```python
import os
import shutil
import random

# Define dataset paths
dataset_path = "/kaggle/working/Foul-4"
train_images_path = os.path.join(dataset_path, "train/images")
train_labels_path = os.path.join(dataset_path, "train/labels")
valid_images_path = os.path.join(dataset_path, "valid/images")
valid_labels_path = os.path.join(dataset_path, "valid/labels")

# Ensure validation folders exist
os.makedirs(valid_images_path, exist_ok=True)
os.makedirs(valid_labels_path, exist_ok=True)

# Get list of images in the training set
images = [f for f in os.listdir(train_images_path) if f.endswith((".jpg", ".png", ".jpeg"))]

# Calculate 10% of images to move
num_to_move = int(len(images) * 0.1)
selected_images = random.sample(images, num_to_move)

# Move images and their labels
for img in selected_images:
    label = img.rsplit(".", 1)[0] + ".txt"  # Match corresponding label file

    # Move image
    shutil.move(os.path.join(train_images_path, img), os.path.join(valid_images_path, img))

    # Move label if it exists
    label_path = os.path.join(train_labels_path, label)
    if os.path.exists(label_path):
        shutil.move(label_path, os.path.join(valid_labels_path, label))

print(f"Moved {num_to_move} images and their labels to the validation folder.")
```
